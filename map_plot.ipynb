```python
"""""""""""""""""""""""""""""""""""""""""""""""
    ||       Program by: Hang Chen Xiong  ||
    ||       ID: 753057                   ||
    ||       Date: May 2016               ||
    ||       Project - COMP20008          ||
"""""""""""""""""""""""""""""""""""""""""""""""
import csv
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import shapefile as shp

%matplotlib inline 
""" 
Map Creation :
    - Using Basemap and coordinates to locate the position of Melbourne on earth.
    - Map image is out sourced from ArcGIS service 
"""
my_map = Basemap(projection='merc', lat_0=-37.8, lon_0=144.9,
    resolution = 'i', area_thresh = 0.1,
    llcrnrlon=144.50, llcrnrlat=-38.20,
    urcrnrlon=145.50, urcrnrlat=-37.50,
    epsg=3577)

# Set resolution
fig = plt.figure(figsize=(16.0, 9.0))
# Sourcing map image
my_map.arcgisimage(service='Canvas/World_Dark_Gray_Base', xpixels = 2000, verbose= True)

""" 
Plotting School locations on map :
    - Coordinate data for all school location in Melbourne is imported from CSV file.
    - Plot point using scatter methods.
"""
# Import csv for school locations
with open('allschoolslist2016.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    X_cord = []
    Y_cord = []
    # Ignore the header row.
    next(reader)
    # Retrieve 'X,Y' coordinates = 'longitude and latitude'
    for var in reader:
        X_cord = X_cord + [var[-2]]#create list of 'x coordinates'
        Y_cord = Y_cord + [var[-1]]#create list of 'y coordinates'

# Combination of both list        
sch_locs = []
sch_locs = zip(X_cord,Y_cord)

# Transform string to float values
for each in sch_locs:
    lons = float(each[0])
    lats = float(each[1])
    # Plot color points on map
    x,y = my_map(lons, lats)
    sch_plot = my_map.scatter(x, y,marker='o', lw=.2, s=12,
            facecolor='#adc1eb', edgecolor='grey',
            alpha=0.9, antialiased=True,
            label='School Locations', zorder=3)
    
""" 
Plotting library locations on map :
    - Similar as school location plotting.
"""
# Import csv for library locations
with open('libraries.csv', 'r') as lbry_csvfile:
    lbry_reader = csv.reader(lbry_csvfile)
    lbry_X_cord = []
    lbry_Y_cord = []
    # Ignore the header row.
    next(lbry_reader)
    # Retrieve 'X,Y' coordinates = 'longitude and latitude'
    for lbry_var in lbry_reader:
        lbry_X_cord = lbry_X_cord + [lbry_var[1]]#create list of 'x coordinates'
        lbry_Y_cord = lbry_Y_cord + [lbry_var[0]]#create list of 'y coordinates'

# Combination of both list   
lbry_locs = []
lbry_locs = zip(lbry_X_cord,lbry_Y_cord)

# Transform string to float values
for lbry_each in lbry_locs:
    lbry_lons = float(lbry_each[0])
    lbry_lats = float(lbry_each[1])
    # Plot color points on map
    lbry_x,lbry_y = my_map(lbry_lons, lbry_lats)
    lbry_plot = my_map.scatter(lbry_x, lbry_y,marker='o', lw=.2, s=12,
            facecolor='#ff8566', edgecolor='grey',
            alpha=0.9, antialiased=True,
            label='Library Locations', zorder=3)

#plt.title("Education Facility Locations, Melbourne")
plt.legend([sch_plot, lbry_plot], ['School', 'Library'], prop={'size':8})
plt.show()
```
